```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import *
```

```python
df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
        .load("/Volumes/pyspark/source/rawsource/sales_data_first.csv")
```

```python
df.display()
```

```python
df = df.withColumn("conditional_column", when(col("Units_Sold") == 1, "low")\
    .when(col("Units_Sold") == 2, "medium")\
    .otherwise("high"))
```

```python
df.display()
```

####Using python

```python
def condtional_col(x):

    if (x == 1):
        return "Low"
    elif (x ==2):
        return "Medium"
    else:
        return "High"
```

```python
conditional_col = udf(condtional_col)
```

```python
df = df.withColumn("Conditional_col_new", conditional_col("Units_Sold"))

df.display()
```

####Cumulative_sum

```python
df = df.drop("conditional_column")

df.display()
```

```python
df.withColumn("Cum_sum", sum("Units_Sold").over(Window.partitionBy("Month").orderBy("Month").rowsBetween(Window.unboundedPreceding, Window.currentRow))).display()
```
