```python
%sql

create table pyspark.source.deltasource
(
  id int,
  name string,
  salary int
)
using delta
```

```python
%sql

alter table pyspark.source.deltasource set tblproperties('delta.enableDeletionVectors' = false)
```

```python
%sql
insert into pyspark.source.deltasource values (1, 'John', 100000), (2, 'Mary', 200000), (3, 'Mike', 300000), (4, 'Jane', 400000),(5, 'Mary', 500000), (6, 'Mike', 600000), (7, 'Jane', 700000)
```

```python
%sql
select * from pyspark.source.deltasource
```

```python
df = spark.readStream.option("startingVersion",10).table("pyspark.source.deltasource")
```

```python
df_new = spark.readStream.option("startingVersion",10).table("pyspark.source.deltasource")

df_new.writeStream.format('delta')\
    .option("checkpointLocation", "/Volumes/pyspark/source/sink1_new/checkpoint")\
    .option("path","/Volumes/pyspark/source/sink1_new" )\
    .trigger(once=True)\
        .start()

```

```python
df.writeStream.format('delta')\
    .option("checkpointLocation", "/Volumes/pyspark/source/sink1/checkpoint")\
    .option("path","/Volumes/pyspark/source/sink1" )\
    .trigger(once=True)\
        .start()
    
```
