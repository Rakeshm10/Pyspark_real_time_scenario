```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
        .load("/Volumes/pyspark/source/rawsource/sales_data_first.csv")

df.display()
```

```python
var_Units_Sold = [1,2,3]
```

```python
for i in var_Units_Sold:

    df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
        .load("/Volumes/pyspark/source/rawsource/sales_data_first.csv")\
            .filter(col("Units_Sold") == i)
    
    df.write.format("csv")\
        .mode("append")\
            .option("path",f"/Volumes/pyspark/source/rawsource/loopdata/Units_Sold_{i}")\
                .save()
```
