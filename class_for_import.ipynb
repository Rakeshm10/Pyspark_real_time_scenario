```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import *
```

```python
class window_func:

    df = spark.createDataFrame([(1, 100), (1, 200), (2, 300), (2, 400), (2, 500)], ["id", "v"])
    
    def dense_rank_fun(self,col_name,partition_cols,ord_clos):
        self.df = self.df.withColumn(col_name, dense_rank().over(Window.partitionBy(partition_cols).orderBy(ord_clos)))
        return self.df
    
    def rank_fun(self,col_name,partition_cols,ord_clos):
        self.df = self.df.withColumn(col_name, rank().over(Window.partitionBy(partition_cols).orderBy(ord_clos)))
        return self.df
    
    def row_number_fun(self,col_name,partition_cols,ord_clos):
        self.df = self.df.withColumn(col_name, row_number().over(Window.partitionBy(partition_cols).orderBy(ord_clos)))
        return self.df
    
```
